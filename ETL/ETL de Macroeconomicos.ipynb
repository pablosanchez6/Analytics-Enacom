```python
import csv
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv(r"Dolar Blue.csv",sep=";",header=0)

# Renombrar las columnas
new_column_names = {
    "Unnamed: 0": "Año",
    'ENE': 1,
    'FEB': 2,
    'MAR': 3,
    'ABR': 4,
    'MAY': 5,
    'JUN': 6,
    'JUL': 7,
    'AGO': 8,
    'SEP': 9,
    'OCT': 10,
    'NOV': 11,
    'DIC': 12
}

df = df.rename(columns=new_column_names)
df
```

```python
# Recorrer cada celda con loc
lista = []
for i in range(len(df.index)):
    for j in range(1,len(df.columns)):
        lista.append(df.loc[i, j])
 
print(len(lista))
```

```python
promedios = []
grupo = []

for i in range(len(lista)):
    grupo.append(lista[i])
    if len(grupo) == 3:
        promedios.append(sum(grupo) / 3)
        grupo = []

# Si queda un grupo incompleto, calcular su promedio
if len(grupo) > 0:
    promedios.append(sum(grupo) / len(grupo))

print(promedios)
```

```python
df4= df4.drop(columns=["Periodo","IPC US"])
```

```python
df4= df4.drop(columns=["USDARS oficial","USDARS blue"])
```

```python
df4['Ingresos Dolar Oficial'] = df4['Ingresos (miles de pesos)'] / df4['Dolar Oficial']
df4['Ingresos Dolar Blue'] = df4['Ingresos (miles de pesos)'] / df4['Dolar Blue']
df4['PBI dolar Oficial'] = df4['PBI millones'] / df4['Dolar Oficial']
df4['PBI dola Blue'] = df4['PBI millones'] / df4['Dolar Blue']
```

```python
df4 = df4.drop(df4[(df4['Año'] == 2011) | (df4['Año'] == 2012) | (df4['Año'] == 2023)].index)
```

```python
df4 = df4.round(0)
```

```python
df4.to_csv("Ingresos+Macroeconomia.csv",index=False)
```
